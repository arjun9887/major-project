```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
import numpy as np
import seaborn as sns
sns.set_palette('colorblind')
sns.set(font_scale=1.25, rc={'text.usetex': True})
dark_mode = False
if dark_mode:
    plt.style.use("dark_background")
```

# Teaser

```python
fig, ax = plt.subplots()
ax.set_xlabel('Frames per second [Hz]', fontsize=12)
# ax.set_ylabel('Mean $\mathcal{J}\&\mathcal{F}$ [\%]', fontsize=12)
ax.set_ylabel('$\mathcal{J}$ [\%]', fontsize=12)
ax.tick_params(labelsize=10)
ax.tick_params(axis="x", direction="in", pad=5)
ax.tick_params(axis="y", direction="in", pad=3)

ax.set_xscale('log')

x_ft = [1 / 70, 1 / 13, 1 / 4.5, 1 / 0.43, 1 / 10.0, 1 / 108]
# y_ft = [78.2, 65.3, 68.0, 56.5, 58.6, 70.7]
y_ft = [73.9, 61.6, 64.7, 56.3, 56.6, 67.2]
n_ft = ['PReMVOS', 'OnAVOS', 'OSVOS-S', 'MVOS', 'OSVOS', 'CINM']

xytext_upper_right = (5.0, 0.0)
xytext_lower_right = (-20.0, -12.0)

ax.scatter(x_ft, y_ft, marker='x', c='orangered', label='One-shot fine-tuning')
for i, txt in enumerate(n_ft):
    xytext = xytext_upper_right
    ax.annotate(txt, (x_ft[i], y_ft[i]), xytext=xytext, textcoords='offset points', fontsize=10)

x_no_ft = [1 / 0.51, 1 / 73.6, 1 / (33 / 1000), 1 / 0.3, 1 / 1.8, 1 / 0.16]
# y_no_ft = [71.5, 76.1, 65.7, 64.8, 54.6, 81.6]
y_no_ft = [69.1, 73.4, 63.2, 64.8, 54.6, 79.2]
n_no_ft = ['FEELVOS', 'MHP-VOS', 'RANet', 'RGMP', 'FAVOS', 'STM']

ax.scatter(x_no_ft, y_no_ft, marker='.', c='cornflowerblue', label='Other methods')
for i, txt in enumerate(n_no_ft):
    xytext = xytext_upper_right
    if txt in ['FEELVOS', 'MHP-VOS']:
        xytext = xytext_lower_right
    ax.annotate(txt, (x_no_ft[i], y_no_ft[i]), xytext=xytext, textcoords='offset points', fontsize=10)
    
# e-OSVOS

ax.scatter([1 / 0.11], [69.0], marker='x', c='orangered')
ax.annotate(r'\textbf{e-OSVOS-10}', (1 / 0.11, 69.0),
            xytext=xytext_upper_right, textcoords='offset points', fontsize=10)

ax.scatter([1 / 0.35], [70.2], marker='x', c='orangered')
ax.annotate(r'\textbf{e-OSVOS-50}', (1 / 0.35, 70.2),
            xytext=xytext_upper_right, textcoords='offset points', fontsize=10)

ax.scatter([1 / 1.05], [72.8], marker='x', c='orangered')
ax.annotate(r'\textbf{e-OSVOS-50-OnA}', (1 / 1.05, 72.8),
            xytext=xytext_upper_right, textcoords='offset points', fontsize=10)

ax.scatter([0.4], [74.4], marker='x', c='orangered')
ax.annotate(r'\textbf{e-OSVOS-100-OnA}', (0.4, 74.4),
            xytext=xytext_upper_right, textcoords='offset points', fontsize=10)

# ax.scatter([1 / 6.47], [76.9], marker='x', c='orangered')
# ax.annotate(r'\textbf{e-OSVOS-50-OnA-50}', (1 / 6.47, 76.9),
#             xytext=(5.0, 5.0), textcoords='offset points', fontsize=10)


x = x_no_ft + x_ft
y = y_no_ft + y_ft

# x += [1 / 0.11]
# y += [60.0]
ax.set_xlim([min(x) - min(x) * 0.4 , max(x) + max(x) * 2.0])
# ax.set_xlim([-0.35 , max(x) + max(x) * 0.35])
ax.set_ylim([min(y) - 2, max(y) + 2])
ax.legend(loc='lower left', fontsize=8)

if dark_mode:
    plt.grid(True, which="both", color='dimgrey')
else:
    plt.grid(True, which="both")
    
sns.despine(left=True, bottom=True, right=True)

plt.savefig('teaser.pdf', format='pdf', bbox_inches='tight')
plt.show()
```
